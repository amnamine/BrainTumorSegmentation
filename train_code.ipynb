```python
import os
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split

# ==========================================
# 1. CONFIGURATION & MULTI-GPU SETUP
# ==========================================
# Paths - Adjusting to Kaggle's typical mount structure
BASE_DIR_1 = '/kaggle/input/datasets/nikhilroxtomar/brain-tumor-segmentation'
BASE_DIR_2 = '/kaggle/input/brain-tumor-segmentation'
BASE_DIR = BASE_DIR_1 if os.path.exists(BASE_DIR_1) else BASE_DIR_2

IMAGE_DIR = os.path.join(BASE_DIR, 'images')
MASK_DIR = os.path.join(BASE_DIR, 'masks')

IMG_SIZE = 256
EPOCHS = 40

# Initialize Multi-GPU Strategy for T4x2
strategy = tf.distribute.MirroredStrategy()
print(f'Number of GPUs in sync: {strategy.num_replicas_in_sync}')

GLOBAL_BATCH_SIZE = 32 * strategy.num_replicas_in_sync

# ==========================================
# 2. DATA PREPARATION PIPELINE & AUGMENTATION
# ==========================================
image_paths = sorted([os.path.join(IMAGE_DIR, f) for f in os.listdir(IMAGE_DIR) if f.endswith('.png')])
mask_paths = sorted([os.path.join(MASK_DIR, f) for f in os.listdir(MASK_DIR) if f.endswith('.png')])

train_img_paths, val_img_paths, train_mask_paths, val_mask_paths = train_test_split(
    image_paths, mask_paths, test_size=0.15, random_state=42
)

def process_path(img_path, mask_path):
    # Read and decode image
    img = tf.io.read_file(img_path)
    img = tf.image.decode_png(img, channels=3)
    img = tf.image.resize(img, [IMG_SIZE, IMG_SIZE])
    img = tf.cast(img, tf.float32) / 255.0

    # Read and decode mask
    mask = tf.io.read_file(mask_path)
    mask = tf.image.decode_png(mask, channels=1)
    mask = tf.image.resize(mask, [IMG_SIZE, IMG_SIZE])
    mask = tf.cast(mask, tf.float32) / 255.0
    mask = tf.cast(mask > 0.5, tf.float32) # Strictly binarize

    return img, mask

def augment(img, mask):
    # Concatenate to apply exact same random transformations to both image and mask
    combined = tf.concat([img, mask], axis=-1)
    combined = tf.image.random_flip_left_right(combined)
    combined = tf.image.random_flip_up_down(combined)

    img = combined[:, :, :3]
    mask = combined[:, :, 3:]
    return img, mask

def build_dataset(img_paths, mask_paths, is_train=True):
    ds = tf.data.Dataset.from_tensor_slices((img_paths, mask_paths))
    ds = ds.map(process_path, num_parallel_calls=tf.data.AUTOTUNE)
    if is_train:
        ds = ds.map(augment, num_parallel_calls=tf.data.AUTOTUNE)
        ds = ds.shuffle(1000)
    ds = ds.batch(GLOBAL_BATCH_SIZE)
    ds = ds.prefetch(tf.data.AUTOTUNE)
    return ds

train_dataset = build_dataset(train_img_paths, train_mask_paths, is_train=True)
val_dataset = build_dataset(val_img_paths, val_mask_paths, is_train=False)

# ==========================================
# 3. METRICS, LOSS & U-NET MODEL
# ==========================================
def dice_coef(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    return (2. * intersection + smooth) / (tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) + smooth)

def iou(y_true, y_pred, smooth=1e-6):
    y_true_f = tf.reshape(y_true, [-1])
    y_pred_f = tf.reshape(y_pred, [-1])
    intersection = tf.reduce_sum(y_true_f * y_pred_f)
    union = tf.reduce_sum(y_true_f) + tf.reduce_sum(y_pred_f) - intersection
    return (intersection + smooth) / (union + smooth)

# Crucial fix: Combined loss for class imbalance
def bce_dice_loss(y_true, y_pred):
    bce = tf.keras.losses.binary_crossentropy(y_true, y_pred)
    dice_loss = 1.0 - dice_coef(y_true, y_pred)
    return bce + dice_loss

def conv_block(inputs, filters, dropout_rate=0.0):
    x = tf.keras.layers.Conv2D(filters, 3, padding="same", kernel_initializer="he_normal")(inputs)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)

    if dropout_rate > 0:
        x = tf.keras.layers.Dropout(dropout_rate)(x)

    x = tf.keras.layers.Conv2D(filters, 3, padding="same", kernel_initializer="he_normal")(x)
    x = tf.keras.layers.BatchNormalization()(x)
    x = tf.keras.layers.Activation("relu")(x)
    return x

def build_unet(input_shape):
    inputs = tf.keras.layers.Input(input_shape)

    # Encoder
    c1 = conv_block(inputs, 64)
    p1 = tf.keras.layers.MaxPooling2D(2)(c1)

    c2 = conv_block(p1, 128)
    p2 = tf.keras.layers.MaxPooling2D(2)(c2)

    c3 = conv_block(p2, 256)
    p3 = tf.keras.layers.MaxPooling2D(2)(c3)

    c4 = conv_block(p3, 512, dropout_rate=0.3) # Added dropout to prevent overfitting
    p4 = tf.keras.layers.MaxPooling2D(2)(c4)

    # Bridge
    b1 = conv_block(p4, 1024, dropout_rate=0.3)

    # Decoder
    u1 = tf.keras.layers.Conv2DTranspose(512, 2, strides=2, padding="same")(b1)
    u1 = tf.keras.layers.concatenate([u1, c4])
    c5 = conv_block(u1, 512)

    u2 = tf.keras.layers.Conv2DTranspose(256, 2, strides=2, padding="same")(c5)
    u2 = tf.keras.layers.concatenate([u2, c3])
    c6 = conv_block(u2, 256)

    u3 = tf.keras.layers.Conv2DTranspose(128, 2, strides=2, padding="same")(c6)
    u3 = tf.keras.layers.concatenate([u3, c2])
    c7 = conv_block(u3, 128)

    u4 = tf.keras.layers.Conv2DTranspose(64, 2, strides=2, padding="same")(c7)
    u4 = tf.keras.layers.concatenate([u4, c1])
    c8 = conv_block(u4, 64)

    outputs = tf.keras.layers.Conv2D(1, 1, padding="same", activation="sigmoid")(c8)
    return tf.keras.models.Model(inputs, outputs)

# Build and Compile within Strategy Scope
with strategy.scope():
    model = build_unet((IMG_SIZE, IMG_SIZE, 3))
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-3),
                  loss=bce_dice_loss, # Using new robust loss
                  metrics=[dice_coef, iou]) # Removed misleading 'accuracy'

# ==========================================
# 4. TRAINING
# ==========================================
callbacks = [
    # Saving in native Keras format and monitoring the correct metric
    tf.keras.callbacks.ModelCheckpoint('unet_brain_tumor.keras', monitor='val_dice_coef', mode='max', verbose=1, save_best_only=True),
    tf.keras.callbacks.ReduceLROnPlateau(monitor='val_dice_coef', mode='max', factor=0.5, patience=4, verbose=1),
    tf.keras.callbacks.EarlyStopping(monitor='val_dice_coef', mode='max', patience=10, restore_best_weights=True)
]

print("Starting training...")
history = model.fit(
    train_dataset,
    validation_data=val_dataset,
    epochs=EPOCHS,
    callbacks=callbacks
)

# ==========================================
# 5. EVALUATION & CURVES
# ==========================================
plt.figure(figsize=(16, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['dice_coef'], label='Train Dice')
plt.plot(history.history['val_dice_coef'], label='Val Dice')
plt.title('Model Dice Coefficient')
plt.xlabel('Epochs')
plt.ylabel('Dice Score')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Train Loss (BCE+Dice)')
plt.plot(history.history['val_loss'], label='Val Loss (BCE+Dice)')
plt.title('Model Combined Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()

# ==========================================
# 6. VISUALIZATION ON 10 RANDOM IMAGES
# ==========================================
print("\nVisualizing 10 Random Test Predictions...")
model.load_weights('unet_brain_tumor.keras')

import random
random_indices = random.sample(range(len(val_img_paths)), 10)

plt.figure(figsize=(15, 30))
for i, idx in enumerate(random_indices):
    img_path = val_img_paths[idx]
    mask_path = val_mask_paths[idx]

    img, true_mask = process_path(img_path, mask_path)

    img_expanded = tf.expand_dims(img, axis=0)
    pred_mask = model.predict(img_expanded, verbose=0)[0]
    pred_mask_bin = (pred_mask > 0.5).astype(np.float32)

    plt.subplot(10, 3, i*3 + 1)
    plt.imshow(img.numpy())
    plt.title("Original Image")
    plt.axis('off')

    plt.subplot(10, 3, i*3 + 2)
    plt.imshow(true_mask.numpy(), cmap='gray')
    plt.title("True Mask")
    plt.axis('off')

    plt.subplot(10, 3, i*3 + 3)
    plt.imshow(pred_mask_bin, cmap='gray')
    plt.title(f"Predicted Mask")
    plt.axis('off')

plt.tight_layout()
plt.show()
print("Execution Complete. Optimized model saved as 'unet_brain_tumor.keras'")
```
